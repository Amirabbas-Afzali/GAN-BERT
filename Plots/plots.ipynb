```python
res10 = np.load('/content/results_G1_10.pickle',allow_pickle=True)
res01 = np.load('/content/results_0.01.pickle',allow_pickle=True)
res05 = np.load('/content/results_G1_05.pickle',allow_pickle=True)
res50 = np.load('/content/results_0.5_G1.pickle',allow_pickle=True)
```

```python
acc10 = []
mcc10 = []
eval_loss10 = []
acc01 = []
mcc01 = []
eval_loss01 = []
acc50 = []
mcc50 = []
eval_loss50 = []

acc05 = []
mcc05 = []
eval_loss05 = []


for res in res10:
  acc10.append(res['acc'])
  mcc10.append(res['mcc'])
  eval_loss10.append(res['eval_loss'])
for res in res01:
  acc01.append(res['acc'])
  mcc01.append(res['mcc'])
  eval_loss01.append(res['eval_loss'])
for res in res50:
  acc50.append(res['acc'])
  mcc50.append(res['mcc'])
  eval_loss50.append(res['eval_loss'])
for res in res05:
  acc05.append(res['acc'])
  mcc05.append(res['mcc'])
  eval_loss05.append(res['eval_loss'])

acc10 = np.array(acc10)
mcc10 = np.array(mcc10)
eval_loss10 = np.array(eval_loss10)

acc01 = np.array(acc01)
mcc01 = np.array(mcc01)
eval_loss01 = np.array(eval_loss01)

acc50 = np.array(acc50)
mcc50 = np.array(mcc50)
eval_loss50 = np.array(eval_loss50)

acc05 = np.array(acc05)
mcc05 = np.array(mcc05)
eval_loss05 = np.array(eval_loss05)
```

```python
epoch = np.array(list(range(1,9)))
plt.plot(epoch,acc50*100,marker='o')
plt.plot(epoch,acc10*100,marker='o')
plt.plot(epoch,acc05*100,marker='o')
plt.plot(epoch,acc01*100,marker='o')

plt.title('accuracy of GAN bert with G1')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['50% labed data','10% labed data','5% labed data','1% labed data'])
plt.grid(True)
```

```python
epoch = np.array(list(range(1,9)))
plt.plot(epoch,mcc50*100,marker='o')
plt.plot(epoch,mcc10*100,marker='o')
plt.plot(epoch,mcc05*100,marker='o')
plt.plot(epoch,mcc01*100,marker='o')

plt.title('mcc of GAN bert with G1')
plt.xlabel('epoch')
plt.ylabel('mcc')
plt.legend(['50% labed data','10% labed data','5% labed data','1% labed data'])
plt.grid(True)
```

```python
epoch = np.array(list(range(1,9)))
plt.plot(epoch,eval_loss50*100,marker='o')
plt.plot(epoch,eval_loss10*100,marker='o')
plt.plot(epoch,eval_loss05*100,marker='o')
plt.plot(epoch,eval_loss01*100,marker='o')

plt.title('loss of GAN bert with G1')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['50% labed data','10% labed data','5% labed data','1% labed data'])
plt.grid(True)
```
